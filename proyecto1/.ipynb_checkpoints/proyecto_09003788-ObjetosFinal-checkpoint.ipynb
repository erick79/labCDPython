### 5. Para cada variable independiente x :
Calcular el coeficiente de correlaci´on entre x y y.<br>
Graficar x vs y(scatterplot) usando matplotlib.<br>
Colocar el coeficiente de correlaci´on y colocarlo como parte del t´ıtulo de la gr´afica.<br>
Basado en la gr´afica y el coeficiente de correlaci´on de cada par x,y elegir las 2 variables<br>
con m´as potencial predictivo es decir las 2 variables que presentan mayor correlaci´on<br>
entre dicha variable y la variable dependiente<br>

```python

# Definimos la variable dependiente 'y'
y = dfEntrenamiento['SalePrice']

# Lista para almacenar los coeficientes de correlación
correlaciones = []

# Iteramos sobre las columnas, excepto 'SalePrice' que es nuestra variable dependiente
for columna in dfEntrenamiento.columns.drop('SalePrice'):
    x = dfEntrenamiento[columna]
    
    # Calculamos el coeficiente de correlación de Pearson
    coef_correlacion = x.corr(y)
    
    # Añadimos el coeficiente a la lista para análisis posterior
    correlaciones.append((columna, coef_correlacion))
    
    # Creamos el scatterplot
    plt.figure(figsize=(6, 3))
    sns.scatterplot(x=x, y=y)
    plt.title(f'{columna} vs SalePrice (Correlación: {coef_correlacion:.2f})')
    plt.xlabel(columna)
    plt.ylabel('SalePrice')
    plt.show()

# Identificamos las 2 variables con mayor coeficiente de correlación
variables_con_mayor_correlacion = sorted(correlaciones, key=lambda x: abs(x[1]), reverse=True)[:2]

print("Las dos variables con mayor potencial predictivo son:")
for variable, correlacion in variables_con_mayor_correlacion:
    print(f"{variable}, con un coeficiente de correlación de {correlacion:.2f}")
```

#### 6.Crear una funcion para entrenar un modelo de regresion lineal de una variable y = β0 +β1 ∗x.
La funcion recibe como argumentos:<br>
6.1 Vector con la variable independiente x,<br>
6.2 Vector con la variable dependiente y,<br>
6.3 un entero epochs que indica por cuantas iteraciones entrenar el modelo.<br>
6.4 un entero imprimir error cada , que nos indica cada cuantas iteraciones queremos imprimir a traves de print: <br>
    el numero de iteracion, el error del modelo en esa iteracion, si imprimir error cada = 10, <br>
    se despliega en pantalla el error en las iteraciones:<br>
    10,20,30,40,50.<br>
6.5 escalar α(learning rate): es usado como parte de la expresion matematica para actualizar<br>
    en cada iteracion los parametros del modelo.<br>

```python
class Regresion():
    #constructor
    def __init__(self, x, y, epochs, imprimir_error_cada,learning_rate):
       #Capturando los parámetros de la regresión y los ponemos como atributos de la clase
        self.imprimir_error_cada = imprimir_error_cada
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.x_original = x
        self.vectorUnos = np.ones_like(x).reshape(-1,1)
        print(self.imprimir_error_cada)
        # Inicialización de parámetros
        self.b0 = np.random.default_rng().normal(np.average(self.x_original), np.std(self.x_original))
        
        self.vectory = np.reshape(y, (-1,1))
        self.b1 = np.random.default_rng().normal(np.average(self.vectory), np.std(self.vectory))
        n = len(y)
        vectorUnos = np.ones_like(x).reshape(-1,1)
        #Crear una matriz de 2 columnas, la primera columna corresponde al vector de datos x y la segunda columna de la matriz para todas las filas es igual a 1
        self.vectorx = np.reshape(x,(-1,1))
        self.vectorx = np.hstack([self.vectorx,self.vectorUnos])
        self.vectory = np.reshape(y, (-1,1))
        print(self.vectorx)
        print(self.vectory)
        #Inicializar los par´ametros del modelo en un vector β0 y β1, esto es equivalente a empezar
        #el proceso con una recta inicial la cual en cada iteraci´on actualizaremos hasta encontrar
        #una que aproxime de buena manera los datos x, y
        self.betas = np.array([[self.b1],[self.b0]])
        self.errors = []
        self.models = {}
        for i in range(0, epochs):
                #Predicciones
                prediccionY = np.matmul(self.vectorx,self.betas)
                
                #Calculo del error
                e = 1/(2*self.vectory.shape[0]) * np.sum((self.vectory - prediccionY)**2)
                self.errors.append(e)
                
                #Creando una estructura de datos (diccionario) que almacene los betas, 
                #la y predicha, la y real, y el valor de x sin un array de 1s
                self.models[i] = [self.betas, prediccionY, self.vectory, self.x_original]
    
                #calculando gradientees
                gradiente = (1/self.vectory.shape[0]) * sum((prediccionY - self.vectory)*self.vectorx)
                matG = (self.learning_rate*gradiente)
                matG = matG.reshape((-1,1))
    
                #Actualizando Betas
                self.betas = self.betas - matG
    
                #Imprimiendo resultados cada n veces
                if i%imprimir_error_cada==0:
                    print('Iteración: ', i)
                    print('Error: ',e)
                    print("")

    #este metodo devuelve 2 resultados: La estructura de datos conteniendo el modelo de
    #cada iteracion y el error de cada iteraci´on   
    def getModelos(self):
        return self.models , np.array(self.errors)
    
    #este metodo es el que vectoriza 
    def predictions(self, x_rp):
        self.x_rp = x_rp
        self.vect_ones_xp = np.ones_like(self.x_rp).reshape(-1,1)
        self.xp = np.reshape(self.x_rp,(-1,1))
        self.xp = np.hstack([self.xp,self.vect_ones_xp])
        self.yp = np.matmul(self.xp, self.betas)
        return self.yp

    #este metodo que permite calcular los errores entre los valores reales de Y, y los valores predichos
    def error_predictions(ypp, ypr):
        ep = 1/(2*ypp.shape[0]) * np.sum((ypr - ypp)**2)
        return ep
    
    '''Crear una funci´on que nos permita visualizar con matplotlib c´omo cambia el error en
    el tiempo: crear una funci´on que tome como par´ametro el vector de errores generados
    por la funci´on de entrenamiento y grafique en el eje x el nu´mero de iteraci´on y en
    el eje y el error para esa iteraci´on 
    '''
    def cambiosError(errors):
        plt.figure(figsize=(8, 6))
        plt.plot(errors, linewidth=2)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
        plt.title("Errores del modelo", fontsize=14)
        plt.xlabel("Epoch", fontsize=12)
        plt.ylabel("Errores", fontsize=12)
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
        plt.show()

    
    #metodo que grafica cada modelo y sus errores
    def graficaErrores(modelos, y):
        i = 1
        labels = []
        e_modelos = []
        for r in modelos:
            e_modelos.append(Regresion.error_predictions(r, y))
            labels.append('Modelo'+str(i) + '\n'+str(round(Regresion.error_predictions(r, y),0)))
            i=i+1
        plt.figure(figsize =(8,5))
        fig, ax = plt.subplots()
        ax.bar(labels, e_modelos)
        fig.tight_layout()
        if (i%3==0):
            plt.show()
    
    '''
    Crear una funci´on que nos permita visualizar con matplotlib c´omo evoluciona el
    modelo entrenado en el tiempo : Crear una funci´on que tome como par´ametro la
    estructura de datos conteniendo el historial de modelos ,y un valor ?n? que indica
    cada cuantas iteraciones graficar el modelo resultante, por ejemplo para n=3 la
    funci´on debe graficar la recta correspondiente al modelo junto a los datos cada 3
    iteraciones, si se ejecutaron 15 iteraciones , para n = 3 se grafica el modelo de las
    iteraciones 3,6,9,12,15 junto con los datos de entrenamiento
    '''
    
    def evolucionModelo(models_t,n):
        fig = plt.figure(figsize = (9,6))
        ax =  fig.add_axes([0.1,0.1,0.7,0.7])
        ax.set_title('Comparativa de modelos')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        i = 0
        for k in models_t.keys():
            if k%n == 0 and k != 0:
                ax.scatter(x = models_t[k][3], y = models_t[k][2], color = '#CDCDCD')
                ax.plot(models_t[k][3],  models_t[k][1], linestyle ='solid', color ='#93b5d9')

            if k == max(models_t.keys()):
                ax.plot(models_t[k][3],  models_t[k][1], linestyle ='solid', color ='#0051ff')
                ax.text(max(models_t[k][3]),max(models_t[k][1])+100, 'Modelo '+str(k), fontsize='small')
        plt.show()
        
    '''
    Crear una funci´on que nos permita visualizar con matplotlib c´omo evoluciona el modelo entrenado en 
    el tiempo : Crear una funci´on que tome como par´ametro la estructura de datos conteniendo el historial 
    de modelos ,y un valor ?n? que indica cada cuantas iteraciones graficar el modelo resultante, 
    por ejemplo para n=3 la funci´on debe graficar la recta correspondiente al modelo junto a los datos 
    cada 3 iteraciones, si se ejecutaron 15 iteraciones , para n = 3 se grafica el modelo de las iteraciones 
    3,6,9,12,15 junto con los datos de entrenamiento
    betas_plot
    '''
    def evolucionModeloEntrenado(models,n1):
        b0 = []
        b1 = []
        for k in models.keys():
            b1.append(models[k][0][0])
            b0.append(models[k][0][1])
        fig = plt.figure(figsize = (8,5))
        ax =  fig.add_axes([0.1,0.1,0.5,0.5])
        ax.plot(b0)
        ax.plot(b1)
        ax.legend(['B0','B1'])
        ax.set_title('Evolución de Betas')
        ax.set_ylabel('Betas')
        ax.set_xlabel('Epocs')
        
    #metodo que compara ambos modelos                
    def comprarAmbosModelos(modelo1,modelo2,x):
        #predicciones del modelo 1 recibido en parametro
        PM1 = modelo1.predictions(x)
        #predicciones del modelo 2 con la libreria
        PM2 = modelo2.predict(x.reshape(-1,1))        
        #los dos modelos
        promedio = (PM1 + PM2)/2
        return PM1, PM2, promedio
    
    
    
```
